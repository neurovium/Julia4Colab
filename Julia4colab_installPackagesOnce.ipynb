Install Julia:

Use a shell command to install Julia if it's not already installed on the Colab instance.

```
# Install Julia
%%capture
%%shell
if ! command -v julia 3>&1 > /dev/null; then
    wget -q 'https://julialang-s3.julialang.org/bin/linux/x64/1.10/julia-1.10.1-linux-x86_64.tar.gz' -O /tmp/julia.tar.gz
    tar -x -f /tmp/julia.tar.gz -C /usr/local --strip-components 1
    rm /tmp/julia.tar.gz
fi
julia -e 'using Pkg; pkg"add IJulia; precompile;"'
echo 'Done'
```

```
# Set Up Configuration Directory and Environment Variable
%%shell
mkdir -p ~/.julia/config # this sets up in the virtual machine
if ! grep -q "ENV[\"JULIA_DEPOT_PATH\"]" ~/.julia/config/startup.jl; then
    echo 'ENV["JULIA_DEPOT_PATH"] = "/content/drive/MyDrive/julia_depot"' >> ~/.julia/config/startup.jl
fi

# Print the contents of the startup.jl to verify
cat ~/.julia/config/startup.jl


# Set Julia to use the Google Drive path for packages
echo 'ENV["JULIA_DEPOT_PATH"] = "/content/drive/MyDrive/julia_depot"' >> ~/.julia/config/startup.jl

# Create the environment directory in Google Drive if it doesn't exist
mkdir -p /content/drive/MyDrive/julia_depot
```

```
# Create julia_depot Directory (This is for package installation destination in google drive)
%%shell
if [ ! -d /content/drive/MyDrive/julia_depot ]; then
    mkdir -p /content/drive/MyDrive/julia_depot
fi
```

```
# Test the Environment Variable in Julia
%%shell
julia -e '
println("Testing JULIA_DEPOT_PATH:");
println(ENV["JULIA_DEPOT_PATH"]);
'
```

Now we install Julia packages. But note that we will set the destination to be google drive.

```
%%shell
export JULIA_DEPOT_PATH="/content/drive/MyDrive/julia_depot"

# Create and activate a new environment in the specified directory
julia --project="/content/drive/MyDrive/JuliaColab" -e '
using Pkg;
Pkg.activate("/content/drive/MyDrive/JuliaColab");
Pkg.add(["DataStructures", "RandomNumbers", "StatsBase"]);
Pkg.instantiate();
Pkg.status();
'
```
